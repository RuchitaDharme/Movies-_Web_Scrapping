Page One

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url_1 = "https://www.themoviedb.org/movie/"
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
source_data = requests.get(url_1,headers = header).text
soup_data = BeautifulSoup(source_data,'lxml')
```

```python
#to find all the moives of first page
movies = soup_data.find_all('div',class_='card style_1')
```

```python
page_1_list=[]

for movie in movies:
    
    base_url = "https://www.themoviedb.org/movie/"
    exp = str(movie.find('a'))[30:38]
    digits = re.findall('\d',exp)
    extension = ''.join(digits)
    M_url = base_url + extension
    
    try:
        m_source_data = requests.get(M_url,headers = header).text
    except:
        m_source_data = requests.get(M_url,headers = header).text
    print(M_url)    
    soup_movie_data = BeautifulSoup(m_source_data,'lxml')
    movie_name = soup_movie_data.find('h2').find('a').text 
    release_date = soup_movie_data.find('span',class_='release').text.replace('\n','').strip()
    genre = soup_movie_data.find('span',class_='genres').text.replace('\n','').replace('\xa0','')
    runtime = soup_movie_data.find('span',class_='runtime').text.replace('\n','').strip()
    rating = float(str(soup_movie_data.find('div',{'class':'user_score_chart'}))[69:73])
    director_info = soup_movie_data.find('li',class_='profile')
    director = director_info.find('a').text
    
    page_1_dict = {
        'Movie Name':movie_name,
        'Release Date':release_date,
        'Genre':genre,
        'Runtime':runtime,
        'Rating': rating,
        'Director': director
    }
    
    page_1_list.append(page_1_dict)
```

```python
print(page_1_list)
```

```python
df = pd.DataFrame(page_1_list)
df
```

```python
df.to_excel('Page1Movie.xlsx')
```
